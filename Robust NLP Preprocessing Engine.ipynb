**Task 2:Build Advanced Preprocessing Function**

```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|\S+@\S+|\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r'\s+', ' ', text).strip()

    return " ".join([
        w for w in text.split()
        if len(w) > 2 or w in ("no", "not")
    ])
```

**Task 4:Token Analysis** 

```python

def preprocess_text(text):

    text = text.lower()

    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)

    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()

    filtered_words = [
        w for w in words if len(w) > 2 or w in ["no", "not"]
    ]

    return filtered_words


```

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
max_noise = 0
most_noise_sentence = ""

max_tokens = 0
most_meaningful_sentence = ""

for i, sentence in enumerate(sentences, 1):

    tokens = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = sum(len(w) for w in tokens) / total_tokens
    else:
        avg_length = 0

    # Noise = original words - cleaned tokens
    original_words = len(sentence.split())
    noise = original_words - total_tokens

    # Track most noisy sentence
    if noise > max_noise:
        max_noise = noise
        most_noise_sentence = sentence

    # Track most meaningful (max tokens after cleaning)
    if total_tokens > max_tokens:
        max_tokens = total_tokens
        most_meaningful_sentence = sentence

    # -------- Output --------
    print(f"\nSentence {i}:")
    print("Original Text        :", sentence)
    print("Tokens               :", tokens)
    print("Total Tokens         :", total_tokens)
    print("Unique Tokens        :", unique_tokens)
    print("Average Token Length :", round(avg_length, 2))
print("\n--- Analysis Results ---")
print("Sentence with most noise        :", most_noise_sentence)
print("Sentence with most meaningful tokens :", most_meaningful_sentence)
```
